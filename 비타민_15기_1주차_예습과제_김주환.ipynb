### 2.1 (5점)

문제: 아래 코드는 iris 데이터셋 분석을 위한 전처리 코드입니다. 아래 보기에서 해당 코드에서 사용된 전처리 과정을 선택하여 나열하시오. 

a) 결측치 처리  
b) 이상치 처리  
c) 정규화(스케일링)  
d) 원핫 인코딩  
e) 특성 선택  
f) 데이터 변환  

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder

def preprocess_data():
    iris = load_iris()
    df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
    
    df.fillna(df.mean(), inplace=True)
    
    scaler = MinMaxScaler()
    num_features = iris.feature_names
    df[num_features] = scaler.fit_transform(df[num_features])
    
    ohe = OneHotEncoder(sparse=False)
    target_encoded = ohe.fit_transform(df[['target']])
    target_encoded_df = pd.DataFrame(target_encoded.toarray(), columns=ohe.get_feature_names_out(['target']))
    df = pd.concat([df.drop('target', axis=1), target_encoded_df], axis=1)
    
    return df
```

정답: a, c, d

### 2.4 (5점)

문제: 다음은 이상치를 계산하고, 제거하는 코드입니다. 흐름에 맞게 ? 부분을 채우시오.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    'Score': [45, 50, 55, 100, 200, 48, 52, 49]
})

# IQR 계산
Q1 = df['Score'].quantile(0.25)
Q3 = df['Score'].quantile(0.75)
IQR = Q3 - Q1

# 이상치 제거
filtered_df = df[~((df['Score'] < (Q1 - 1.5 * IQR)) | (df['Score'] > (Q3 + 1.5 * IQR)))]
print(filtered_df)
```

### 3.2 (5점)

문제: pandas 의 describe 함수를 사용하여 데이터의 기술 통계 값을 확인할 수 있습니다. 주석을 읽고, 요구 조건에 맞게 빈칸을 채우시오.

```python
import seaborn as sns
import pandas as pd

# 타이타닉 데이터셋 로드
data = sns.load_dataset('titanic')

# 모든 변수의 기술 통계 값 출력 (수치형 + 범주형 변수 포함)
print("모든 변수 기술 통계 값 출력")
display(data.describe(include='all'))  # 모든 데이터 타입 포함

# 데이터 타입이 category, float64 인 경우의 기술 통계 값 출력
print("데이터 타입이 category, float64 인 경우의 기술 통계 값 출력")
display(data.describe(include=['category', 'float64']))  # 특정 데이터 타입만 포함

# 수치형 데이터의 기술 통계 값에서 특정 분위수 (20%, 30%, 70%) 출력
print("특정 분위에 해당하는 값 출력")
display(data.describe(percentiles=[0.2, 0.3, 0.7]))  # 20%, 30%, 70% 분위수 설정
```
